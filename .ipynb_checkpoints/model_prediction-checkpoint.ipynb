```python
import tensorflow as tf
import cv2
import numpy as np
import requests
import matplotlib.pyplot as plt
```

```python
# Function to preprocess the image
def preprocess_image(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (200, 50))
    img = img.astype(np.float32)  # Convert to float32
    img = img[..., np.newaxis] # Add channel dimension
    img = img / 255.0
    return img
```

```python
# Function to make predictions using TensorFlow Serving
def predict_image(img_array, api_url):
    # Prepare input data for prediction
    img_array = np.expand_dims(img_array, axis=0)
    input_data = {
        "signature_name": "serving_default",
        "instances": img_array.tolist()
    }

    # Send POST request to the prediction endpoint
    response = requests.post(api_url, json=input_data)
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        predictions = response.json()["predictions"]
        return predictions
    else:
        print("Error:", response.text)
        return None
```

```python
# Path to the saved characters_keys list
CHARACTERS_KEYS_PATH = "./models/captcha-ocr-model/1/assets/characters_keys.txt"

# Load the characters_keys list from the text file
characters_keys = []
with open(CHARACTERS_KEYS_PATH, 'r') as f:
    for line in f:
        characters_keys.append(line.strip())

# Now characters_keys contains the loaded list
print(f'Loaded characters keys: {characters_keys}')
```

```python
# Define the API endpoint URL for TensorFlow Serving
# Replace with your TensorFlow Serving endpoint: {YOUR_SERVING_ENDPOINT}/v1/models/captcha-ocr-model:predict
api_url = 'https://ml-deploy-production.up.railway.app/v1/models/captcha-ocr-model:predict' 

# Get the prediction image
# 2g7nm.png, 3d7bd.png, 488de.png, pmf5w.png, w6ny4.jpg
img_path = "./prediction_data/pmf5w.png"
img = preprocess_image(img_path)

# Make prediction using TensorFlow Serving
predictions = predict_image(img, api_url)

# Process predictions (assuming you have characters_keys defined)
if predictions:
    predicted_chars = ''
    for i in range(len(predictions[0])):
        char_probs = predictions[0][f'char_{i+1}']
        predicted_char = characters_keys[np.argmax(char_probs)]
        predicted_chars += predicted_char
    
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(f'Predicted Characters: {predicted_chars}')
    plt.show()
```
